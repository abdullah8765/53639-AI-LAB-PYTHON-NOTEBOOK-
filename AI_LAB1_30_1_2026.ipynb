```python
def is_prime(n):
    if n < 2:
        return 0
    for j in range(2, int(n**0.5) + 1):
        if n % j == 0:
            return 0
    return 1


count = 1
for i in range(2, 100):
    if is_prime(i) == 1:
        print(count, "\t", i)
        if count >= 20:
            break
        count += 1
```
